```python
from google.colab.patches import cv2_imshow
import cv2
import matplotlib.pyplot as plt
%matplotlib inline 
image = cv2.imread("/content/drive/MyDrive/finaldataset/train/009/8.jpg")
saving_image_drives = "/content/drive/MyDrive/noiseless/train/008/8.jpg"
print('original')
cv2_imshow(image)
cv2.destroyAllWindows()
gauss = cv2.GaussianBlur(image, (7,7), 0)
s1 = cv2.addWeighted(image, 2, gauss, -1, 0)
gauss = cv2.GaussianBlur(s1, (7,7), 0)
s2 = cv2.addWeighted(s1, 2, gauss, -1, 0)
gauss = cv2.GaussianBlur(s2, (7,7), 0)
s3 = cv2.addWeighted(s2, 2, gauss, -1, 0)
gauss = cv2.GaussianBlur(s3, (7,7), 0)
s4 = cv2.addWeighted(s3, 2, gauss, -1, 0)
print('sharped')
cv2_imshow(s4)
bw = cv2.threshold(s4, 127, 255, cv2.THRESH_BINARY)
img=bw[1]
print('binary')
cv2_imshow(img)
blur = cv2.GaussianBlur(img,(5,3), 1) 
mask=cv2.inRange(blur,(0,0,0),(150,150,150))
plt.subplot(121), plt.imshow(mask, cmap='gray'), plt.title('masked')
res = 255 - mask
plt.subplot(122), plt.imshow(res, cmap='gray'), plt.title('result')
#cv2.imwrite(saving_image_drives,res)
```
